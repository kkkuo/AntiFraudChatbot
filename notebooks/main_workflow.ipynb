```python
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

# 建立 LangChain 兼容的嵌入器
embedding_model = HuggingFaceEmbeddings(model_name="paraphrase-multilingual-MiniLM-L12-v2")

# 建立 FAISS vectorstore
vectorstore = FAISS.from_texts(texts, embedding=embedding_model)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain_community.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA

model_id = "ziqingyang/chinese-alpaca-2-7b"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", torch_dtype="auto")

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=512)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA

# 1. 自訂 prompt 模板
custom_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
                根據以下案例，請用清楚且溫和的語氣，協助判斷使用者是否遇到詐騙。請避免亂猜，若無法確定請說明原因。

                【詐騙案例】
                {context}

                【使用者提問】
                {question}

                請提供你的判斷與建議：
                """
)

retriever = vectorstore.as_retriever()

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever
)
```

```python
query = "有人打來說我是詐騙案的被害人，要我提供銀行資料，是詐騙嗎？"
result = qa_chain.run(query)
print("回答：", result)
```
